```python
# 21
# 實作 thread-safe counter（使用 threading）
import threading as th
import time
counter = 0 
lock = th.Lock()

def Count(num,repeat):
    global counter 
    for i in range(repeat):
        with lock:
            counter+=num
            print(f"num={num},counter={counter}")
        time.sleep(0.2)
        
thread1 = th.Thread(target=Count, args=(1,10))
thread2 = th.Thread(target=Count,args=(2,20))
thread1.start()
thread2.start()

thread1.join()
thread2.join()
print("Counter end")

```
