```python
# 1
# 將一個 list 中所有偶數平方後回傳新 list（用 list comprehension）

a = [5,6,7,8]
print(type(a))
print(len(a))

b=[value**2 for value in a if value%2==0]
print(b)
```

```python
# 2
# 給定：
# data = ["apple", "banana", "cherry"]
# 轉成：
# {"apple": 5, "banana": 6, "cherry": 6}
data = ["apple", "banana", "cherry"]
dic = {key:len(key) for key in data}
print(dic)
```

```python
list_A = [1,2,3,4,[1,2,3,[1,2,[23,"dafd",["sdf",112]]]]]
#isinstance(var,type) 是否為某個類型

def Flatten(iList):
    res=[] 
    for element in iList:
        if isinstance(element,list):
            #iterator.extend(interator_B) list合併於後面
            res.extend(Flatten(element))
            print(element)
        else:
            res.append(element)
    return res

print(Flatten(list_A))
```

```python
# 4
# 寫一個函式，統計字串中每個字元出現次數（用 dict）
str_A = "a;dkfj;akdlejq;kd;kfn;kanviojc;kadkfa;dnf;a"
def CharCounter(iStr):
    res = {element:iStr.count(element) for element in iStr}
    return res

print(CharCounter(str_A))
        
```

```python
# 5
# 用 sorted() 對以下資料排序（依照第二個欄位）：
data = [("a", 3), ("b", 1), ("c", 2)]

res = sorted(data,key =lambda x:x[1])
print(res)
```

```python
# 7
# 給一個 list，移除重複但保留順序
data = [1,3,5,2,3,2,3,2]
res = []
for element in data: 
    if element not in res:
        res.append(element)
print(res)
```

```python
# 8
# 用一行寫出：
# [1,2,3,4] → ["1","2","3","4"]
data = [1,2,3,4]
res = [str(element) for element in data]
print(res)
```

```python
# 9
# 寫一個函式，回傳兩個 list 的交集（不用 set）
dataA = [1,2,3,4,5,6]
dataB = [1,3,5,7]
def GetInterSection(iListA,iListB):
    res = [element for element in iListA if element in iListB]
    return res
tmp = GetInterSection(dataA,dataB)
print(tmp)
```

```python
# 10
# 讀取一段文字，找出出現最多次的單字
#data = "kdj;akf;eaf;dkjf;leijfdkj;fadkjf;aefjd;fawefasdaaaaaaaa"
data = input("請輸入字串")
champ = 0
for element in data:
    if data.count(element)>=champ:
        champ = data.count(element)
        champ_value = element

print(f"最多次{champ_value},次數{champ}")
```
