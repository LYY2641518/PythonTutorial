```python
import copy
#原始
lst1= [0,1,["a","b","c"]]
#名稱綁定 
lst2 = lst1
lst1[0] = 2
lst1[2][0]="b"
print(f"lst1:{lst1}")
print(f"lst1 id:{id(lst1)}")
print(f"lst2:{lst2}")
print(f"lst2 id:{id(lst2)}\n")

#淺拷貝
lst4 = copy.copy(lst1)
print(f"lst1:{lst1}")
print(f"lst1 id:{id(lst1)}")
print(f"lst4:{lst4}")
print(f"lst4 id:{id(lst4)}\n")
lst1[0] = 3
lst1[2][0]="c"
print(f"lst1:{lst1}")
print(f"lst1 id:{id(lst1)}")
print(f"lst4:{lst4}")
print(f"lst4 id:{id(lst4)}\n")

#深拷貝
lst3 = copy.deepcopy(lst1)
lst1[0] = 4
lst1[2][0]="d"
print(f"lst1:{lst1}")
print(f"lst1 id:{id(lst1)}")
print(f"lst3:{lst3}")
print(f"lst3 id:{id(lst3)}\n")

#深拷貝的內部物件
print(f"lst1[2]:{lst1[2]}")
print(f"lst1[2] id:{id(lst1[2])}")
print(f"lst3[2]:{lst3[2]}")
print(f"lst3[2] id:{id(lst3[2])}\n")
```

```python
import threading
import time

def _task(iName):
    "文件字串"
    print(f"{iName} start")
    time.sleep(2)
    print(f"{iName} end")

# 建立 Thread
thread1 = threading.Thread(target=_task, args=("Thread-1",))
thread2 = threading.Thread(target=_task, args=("Thread-2",))

# 啟動
thread1.start()
thread2.start()

# 等待結束
thread1.join()
thread2.join()

print("All threads finished")
```

```python
import xml.etree.ElementTree as ET
issue = ET.parse("data/110年4月分析陳情焦點議題前十大-依機關類別.csv.xml")
root = issue.getroot()
record = root.findall("RECORD")

for ele in record:
    for item in ele:
        if item.tag == "陳情最多業務項目":
            print(item.text,end = "")
        if item.tag == "該業務占排名前十大百分比":
            print(item.text)
     
```

```python
import xml.etree.ElementTree as ET
issue = ET.parse("data/110年4月分析陳情焦點議題前十大-依機關類別.csv.xml")
root = issue.getroot()

def ShowTree(iTag,iLevel):
    print("     "*iLevel+iTag.tag)
    for child in iTag:
        ShowTree(child,iLevel+1)
    
#ShowTree(root,1)

#num = root.iterfind("編號") iterfind只找下層
num = root.iter("編號")
for node in num:
    print(node.text)

```
