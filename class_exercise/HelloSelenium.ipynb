```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
chromedriver = webdriver.Chrome()

chromedriver.implicitly_wait(10)

chromedriver.get('https://hahow.in/')
time.sleep(15)
source_page = chromedriver.page_source

with open('output/Hahow.html','w') as f :
    f.write(source_page)
    
h2 = chromedriver.find_elements(By.XPATH,'//h2')
for ele in h2:
    print(ele.text)
chromedriver.close()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
chromedriver = webdriver.Chrome()

chromedriver.implicitly_wait(5)

url = 'https://www.taiwanlottery.com/lotto/result/39_m5'
chromedriver.get(url)
sleep(5)
again = True
while(again):
    num = input("請輸入9碼期別")
    if num.isdigit() and len(num)==9 :
        input_form = chromedriver.find_element(By.XPATH,'//input[contains( @id,"el-id")]')
        input_form.clear()
        input_form.send_keys(num)
        sleep(1)
        search_btn = chromedriver.find_element(By.XPATH,'//*[@id="__nuxt"]/main/div[1]/div[2]/div/nav/div[4]/div/button[1]').click()
        
        sleep(3)
        
        numbers = chromedriver.find_elements(By.XPATH,'//div[@class="ball special-number"]')
        for n in numbers:
            print(n.text)
    else :
        print("數字長度錯誤")
        
    again = input("是否繼續查詢 Y/N？")
    if again == "Y":
        again = True
    else:
        again = False
    
    
    
```
