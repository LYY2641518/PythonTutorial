```python
print(np.zeros((2,3)),end="\n\n")# 全 0
print(np.ones((2,3)),end="\n\n")# 全 1
print(np.eye(3),end="\n\n")# 單位矩陣
print(np.arange(0,10,2),end="\n\n")#從0到10，步長為2
print(np.linspace(0,1,5),end="\n\n")#從0到1，取5個點
```

```python
_arrA=np.array([[1,2,3], [4,5,6]])
print(_arrA)
print(_arrA.shape)# 維度
print(_arrA.ndim)# 幾維
print(_arrA.size)# 總元素數
print(_arrA.dtype)# 型別
_arrB=_arrA.reshape(3,2)# 重新塑形
print(_arrA,_arrB,sep="\n\n")
```

```python
_arrA=np.array([10,20,30,40])

print(_arrA[0])# 10
print(_arrA[-1])# 40
print(_arrA[1:3])# [20 30]
_arrA = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(_arrA[0, 1])     # 2
print(_arrA[:, 1])     # 第2欄
print(_arrA[1, :])     # 第2列
```

```python
import numpy as np
np.random.seed(0) # 設定隨機種子，使得每次執行都得到相同的隨機數
print(np.random.rand(3, 4))
print(np.random.rand(3, 4))

# 從列表中隨機選擇3個元素，不允許重複
a = np.random.choice([1, 2, 3, 4, 5], size=3, replace=False) 
print(np.random.shuffle(a))
```
