```python
"""
    QuickSort 快速排序
    1.選一個元素作為基準(pivot)
    2.遞迴陣列將小於pivot往前大於等於pivot不動
        -> 找到基準的正確位置
        -> 將排序範圍減半
    3.將pivot交換到正確位置
    3.對兩個區域進行同樣步驟

"""
data = [23,2,5,564,44,326,745,23,646,34]
def QuickSort(ar,left=None,right=None):
    left = 0 if left is None else left
    right = len(ar)-1 if right is None else right
    if left < right:
        v = partition(ar,left,right)
        QuickSort(ar,left,v-1)
        QuickSort(ar,v+1,right)
        
    
def partition(ar,left,right):
    pivot = left
    i = left +1
    index = i
    while i<=right:
        if ar[i]<ar[pivot]:
            swap(ar,i,index)
            index+=1
        i+=1
    swap(ar,index-1,pivot)
    return index-1
    
def swap(ar,i,j):
    ar[i],ar[j] = ar[j],ar[i]
    
QuickSort(data)
print(data)
```
