```python
import pandas as pd 
df = pd.DataFrame({ 
    "name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "age": ["25", 30, 35, 40, 45],
})
df2 = pd.DataFrame({
        "number": [0, 1, 2, 3 ]
})

df3 = pd.concat([df, df2], axis=1)
print(df3)
#df3.dropna()
df3 = df3.fillna(0)
print(df3)
df3["age"] = df3["age"].replace("25", 60).astype(int)
print(df3)
print(df3["age"].info())
```
