```python
# data from https://allisonhorst.github.io/palmerpenguins/

import matplotlib.pyplot as plt
import numpy as np

species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 250)

plt.show()
```

```python
#子圖 圖例 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
_x = np.linspace(0,2*np.pi,10)
_y = np.sin(_x)
_z = np.cos(_x)

_bar = plt.subplot(2,2,1)
_bar.bar(_x,_y,label = "bar")
_bar.legend()
_bar.set_title("sin")
_plot = plt.subplot(2,2,2)
_plot.plot(_x,_y,color = "#ff6623",linewidth = 2,linestyle="dashed",marker="h",markersize="5",label="plot")
_plot.plot(_x,_z,label = "cos",color="#2366ff",marker="s")
_plot.axis((0,3.0,2.0,-2.0))
_plot.set_ylim(0,2)
_plot.legend(title="legend",facecolor = "#999999",bbox_to_anchor=(1.4,1.05))#圖例
_scatter = plt.subplot(2,1,2)
_scatter.scatter(_x,_z,color = "#88ff55",linewidth = 2,linestyle="dotted",label="scatter")
plt.title("sin cos")
plt.xlabel("cos")
plt.ylabel("value")
plt.legend(bbox_to_anchor = (2.7,0))
plt.show()
```
