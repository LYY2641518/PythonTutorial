```python
from typing import Any
from bs4 import BeautifulSoup as bs 
import requests,os,urllib
import ssl
import lxml
import re 
import logging
#basicConfig只有第一次有用 使用ipynb造成後續調整無用
logging.basicConfig(level=logging.INFO)
logging.getLogger().setLevel(logging.INFO)
ssl._create_default_https_context = ssl._create_unverified_context
url = r"https://www.edu.tw/Default.aspx"
html = requests.get(url)
html.encoding = "utf-8"
soup = bs(html.text,'lxml')
pics_dir = "pics"
if not os.path.exists(pics_dir):
    os.mkdir(pics_dir)
    
imgs = soup.find_all("img")
print(len(imgs))

raw_url = 'https://www.edu.tw'
pic_name = ""   
pic_path = ""
for u in imgs:
    s= u.get("src")
    logging.debug(f"raw src = {s}")
    pic_name = s.split('/')[-1]
    logging.debug(f"{pic_name}")
    if os.path.exists(os.path.join(pics_dir,pic_name)):
        continue
    if pic_name.find(".")<0:
        logging.warning(f"skip pic_name = {pic_name}")
        continue
    #https 直接是圖片
    if "https" in s:
        pic_data: bytes | Any = requests.get(s).content
        logging.info(f"full download path : {s}")
    #/images
    elif "images" in s:
        pic_data: bytes | Any = requests.get(raw_url+u.text).content
        logging.info(f"full download path : {raw_url+u.text}")
        
    with open(os.path.join(pics_dir,pic_name),'wb') as f :
        f.write(pic_data)
        logging.info(f"download pic : {pic_name}")
```
