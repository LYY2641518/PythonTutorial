```python
#透過Nonlocal檢視物件
def Outer():
    num = 1
    print(num)
    print(id(num))
    def Inner(iNum):
        print("Inner before\n")
        print(iNum)
        print(id(iNum))
        iNum = iNum+1
        print("Inner after\n")
        print(iNum)
        print(id(iNum))
    Inner(num)
    print("\n")
    print(num)
    print(id(num))
    def Nonlocal():
        nonlocal num
        print("Nonlocal\n")
        print(num)
        print(id(num))
        
        num = num+1
        print("Nonlocal after\n")
        print(num)
        print(id(num))

    Nonlocal()
    print("\n")
    print(num)
    print(id(num))

Outer()   
```

```python
list_A = [1,2,3,4,[1,2,3,[1,2,[23,"dafd",["sdf",112]]]]]
#isinstance(var,type) 是否為某個類型

def Flatten(iList):
    res=[] 
    for element in iList:
        if isinstance(element,list):
            #iterator.extend(interator_B) list合併於後面
            res.extend(Flatten(element))
            print(element)
        else:
            res.append(element)
    return res

print(Flatten(list_A))
```

```python
class Dog:
    # __init__ 建構式
    def __init__(self, iName):
        self.m_Name = iName
    # __str__ print輸出設定
    def __str__(self):
        return f"Dog: {self.m_Name}"
    # __repr__ debug輸出
    def __repr__(self):
        return f"Dog('{self.m_Name}')"
    
dog = Dog("小黑")

print(dog)        # __str__
print(repr(dog))  # __repr__

ans = getattr(dog,"m_Name")
ans2 = getattr(dog,"__str__")
print(ans)
print(ans2)
print(help(getattr))
```

```python
import time
start_time = time.time()
c = compile("num","string","eval")
for num in range(1000000):
    eval(c)
print(f"compile：{time.time()-start_time}")


start_time = time.time()
for num in range(1000000):
    eval("num")
print(f"eval：{time.time()-start_time}") 
# 👉 原因：
# ✔ eval("num")
# 每次都會：
# parse 字串
# 建 AST
# compile 成 bytecode
# 執行

# ✔ eval(c)
# 只做：
# 直接執行 bytecode
# 👉 少了「解析 + 編譯」 → 所以快很多
```

```python
#貪食蛇
from IPython.display import clear_output
import time
import random
m_Dir_WASD=["w","a","s","d"]
m_Snack = [(5,5),(5,6),(5,7)]
m_Food = [(3,5)]
m_Wall =(
    [(i, 20) for i in range(20)] +      # 上邊
    [(20, i) for i in range(20)] +      # 右邊
    [(i, -1) for i in range(20)] +      # 下邊
    [(-1, i) for i in range(20)] )       # 左邊

while(True):
    for i in range(20):
        print( "",sep="",end="\n")
        for j in range(20):
            if (j,i) in m_Snack:
                print( "\033[37;41m□\033[0m",sep="",end="")
            elif(j,i)in m_Food:
                print( "\033[32m@\033[0m",sep="",end="")
            else :
                print( "■",sep="",end="")
    
    keyinput = input("方向")
    #結束遊戲
    if keyinput=="b":
        break
    elif keyinput in m_Dir_WASD:
        if keyinput =="w":
            newpos = (m_Snack[0][0],m_Snack[0][1]-1)
        elif keyinput =="a":
            newpos = (m_Snack[0][0]-1,m_Snack[0][1])
        elif keyinput =="s":
            newpos = (m_Snack[0][0],m_Snack[0][1]+1)
        elif keyinput =="d":
            newpos = (m_Snack[0][0]+1,m_Snack[0][1])
        else:
             continue
        if  newpos not in m_Snack:
            m_Snack.insert(0,newpos)
            if  newpos in m_Food:
                m_Food.pop()
                def checkfood(x,y):
                    if (x,y) not in m_Snack:
                        return True
                    return False
                new_food = (random.randint(0,19),random.randint(0,19))
                while(True):
                    if checkfood(new_food[0],new_food[1]):
                        break
                    else:
                        new_food = (random.randint(0,19),random.randint(0,19))
                m_Food.insert(0,new_food)
            else:
                m_Snack.pop()

        
    clear_output(wait=True)
    time.sleep(1)
    

```
