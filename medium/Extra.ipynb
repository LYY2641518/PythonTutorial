```python
# 裝飾器 (decorator)
from functools import wraps
import time
#裝飾器 執行時間
def decorator_timing(func):
    @wraps(func)
    def wrapper(*arg,**kwarg):
        print(f"start time:{time.time()}")
        res = func(*arg,**kwarg)
        print(f"end time:{time.time()}")
        return res
    return wrapper
#裝飾器 加倍結果
def decorator_double_result(func):
    #保留原函式的 __name__ 與 __doc__
    @wraps(func) 
    def wrapper(*arg,**kwarg):
        res = func(*arg,**kwarg)
        return res*2
    return wrapper

#套用裝飾器
#@decorator_timing
#@decorator_double_result
def add(a, b):
    return a + b

#等價於 函式重新綁定 
add = decorator_timing(add)
print(add(5,8))

#裝飾器 有參數的
#@decorator等價於 decorator(func)
#@decorator(3) 有參數會等價於 decorator(3)(func) 所以要多包一層
def decorator_repeater(n):
    def decorator_print_number(func):
        def wrapper(*arg,**kwarg):
            res = func(*arg,**kwarg)
            for i in range(n):
                print(i)  
            return res
        return wrapper
    return decorator_print_number
@decorator_repeater(3)
def add2(a,b):
    return a+b

add2(1,3)
```

```python
#迭代器 iterator
class int_stack:
    def __init__(self,*arg):
        self.m_len = 0
        self.m_pos = 0
        self.m_data = []
        self.m_data.append(*arg)
        self.m_len = len(self.m_data)
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.m_pos<self.m_len:
            res = self.m_data[self.m_pos]
            self.m_pos += 1
            return res
        raise StopIteration

st = int_stack([1,3,5])
for i in st:
    print(i)
```
