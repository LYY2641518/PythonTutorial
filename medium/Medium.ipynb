```python
# 11
# 寫一個 decorator，計算函式執行時間
import time
def decorator_timing(func):
    def wrapper(*arg,**kwarg):
        print(time.time())
        res = func(*arg,**kwarg)
        print(time.time())
        return res
    return wrapper

@decorator_timing
def add(x,y):
    return x+y

add(1,2)
```

```python
# 12
# 實作 generator：
# def fibonacci(n):
# 逐個 yield Fibonacci

def fibonacci(n:int):
    """n目標層數"""
    pre = 1
    pre_pre = 0
    for i in range(n):
        yield pre_pre
        pre ,pre_pre = pre+pre_pre ,pre
        
        
gen = fibonacci(15)
for ele in gen:
    print(ele)

    
```

```python
# 16
# 用 with 自訂 context manager（例如：計時器）
import time
class InputTimer:
    def SetValue(self,input_str):
        self.mInput = int(input_str)
    def __enter__(self):
        print(f"input start time:{time.time()}")
        return self # with的重點
    def __exit__(self, exc_type, exc, tb):
        if exc_type :
            print(exc_type)
            print(exc)
            print(tb)
        else:
            print(f"input:{self.mInput}exit time:{time.time()}")
        return True #範例 先吃掉錯誤
with InputTimer() as In:
    In.SetValue(input("請輸入數字"))
    
```

```python
# 18
# 寫一個 retry decorator（失敗重試 N 次）
import random
from functools import wraps
def decorator_retry(n):
    def decorator_retry_fun(iFunc):
        @wraps(iFunc)
        def wrapper(*args,**kwargs):
            for i in range(n):
                try:
                    _res = iFunc(*args,**kwargs)
                    print(f"第{i}次成功")
                    return _res                       
                except Exception as e:
                    print(f"第{i}次失敗,錯誤{e}重試")

        return wrapper
    return decorator_retry_fun 

@decorator_retry(5)                
def show()->bool:
    res = random.randint(1,5)
    print(f"隨機數字為 {res}")

show()
```
